```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Read in the raw data and look at it.
imprisonment_rates_file = "../imprisonment_rates_of_18_24_yos_2000_2012.csv"

imprisionment_rates_df = pd.read_csv(imprisonment_rates_file)
imprisionment_rates_df
imprisionment_rates_df.head(15)
```

```python
# Remove all rows with NA
#imprisionment_rates_with_race_df = imprisionment_rates_df.dropna()
imprisionment_rates_with_race_df = imprisionment_rates_df
imprisionment_rates_with_race_df = imprisionment_rates_with_race_df.drop(['Race/ethnicity'], axis=1)


imprisionment_rates_with_race_df.head(50)
```

```python
#imprisionment_rates_with_race_df['Sex'] = imprisionment_rates_with_race_df['Sex'].fillna("MF")
mymean = imprisionment_rates_with_race_df.groupby(['Year', "Sex"]).mean()
mysem = imprisionment_rates_with_race_df.groupby(['Year', "Sex"]).sem()
#
#mean_df = imprisonment_race_and_totals
#mean_df.head()
#
#
```

```python
# This gets all races and both sexs
try:
    output_data_file = "ImprisonmentRatesWithRace.csv"
    imprisionment_rates_with_race_df.to_csv(output_data_file)
except:
    print(f"Unable to write file: {output_data_file}")
```

```python
# The rows for totals without sex were reoved.
# Get them back.
totals_df = imprisionment_rates_df.iloc[0:13]
totals_df['Sex'] = totals_df['Sex'].fillna("MF")
imprisonment_rates_totals = totals_df.drop(['Race/ethnicity'], axis=1)
imprisonment_rates_totals.head(15)
```

```python
# Save the total data.
try:
    output_data_file = "ImprisonmentRatesTotals.csv"
    imprisonment_rates_totals.to_csv(output_data_file)
except:
    print(f"Unable to write file: {output_data_file}")
```

```python
# Put them both together.
imprisonment_race_and_totals = imprisonment_rates_totals.append(imprisionment_rates_with_race_df)
imprisonment_race_and_totals = imprisonment_race_and_totals.reset_index()
imprisonment_race_and_totals = imprisonment_race_and_totals.drop(['index'], axis=1)
```
